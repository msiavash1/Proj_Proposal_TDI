```python
# I have used this data set with 2017 instances and 16 attributes, among which there are 13 song features attributes 
# and 3 attributes for "song_title", "artist" and "target".

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("spotify_data.csv")
df = df.drop("Unnamed: 0", axis="columns")
df.head()
```

```python
# Pairplot to determine some possible correlated relationships between the 13 attributes.
# This is my first plot for the section 3.
sns.pairplot(df)

```

```python
# I have also used heapmap to review the correlation between the 13 attributes.
# This is my second figure. 
fig2 = plt.figure(figsize=(16, 8))
sns.heatmap(df.corr(), annot=True, annot_kws={'weight':'bold'},linewidths=.5, cmap='YlGnBu')
```

```python
# Here I have calculated the relationship between the acousticness and energy. 
# It can be seen that they have high negative correlation (p=-0.65). 
# Apology for plotting three figures rather than two.
sns.lmplot(y='energy',x='acousticness',data=df, hue='target',palette='BuGn')
```
